```python
# Import Packages
import numpy as np
import cvxpy as cp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_analysis = pd.read_csv('data/complete_data.csv')
unique_brands = df_analysis["brand"].unique()
unique_fuel = df_analysis["fuel_type"].unique()
```

```python
total_fuel = pd.Categorical(df_analysis["fuel_type"], ordered=True)  
fuel_counts = pd.Series(total_fuel).value_counts().sort_index()
plt.figure(figsize=(8,5))
fuel_counts.plot(kind="bar", color = "skyblue", edgecolor="black")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.title("Distribution of Fuel Types in the Dataset")
plt.xticks(rotation=45, ha="right", fontsize=10)  # Reduce font size
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.subplots_adjust(bottom=0.2)  # Add padding to prevent label cutoff
for i, value in enumerate(fuel_counts):
    plt.text(i,value +0.5, str(value), ha="center", va="bottom", fontsize=10)
plt.savefig("data/total_fuel.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
total_import = pd.Categorical(df_analysis["imported"].map({0: 'domestic', 1: 'imported'}))  
import_counts = pd.Series(total_import).value_counts().sort_index()
plt.figure(figsize=(8,5))
import_counts.plot(kind="bar", color = "skyblue", edgecolor="black")
plt.xlabel("Imported/Domestic")
plt.ylabel("Count")
plt.title("Distribution of Imported/Domestic cars in the Dataset")
plt.xticks(rotation=45, ha="right", fontsize=10)  # Reduce font size
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.subplots_adjust(bottom=0.2)  # Add padding to prevent label cutoff
for i, value in enumerate(import_counts):
    plt.text(i,value +0.5, str(value), ha="center", va="bottom", fontsize=10)
plt.savefig("data/total_import.png", dpi=300, bbox_inches="tight")
plt.show()
```
